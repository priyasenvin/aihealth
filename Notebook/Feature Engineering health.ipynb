```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
# Load raw data

df =pd.read_csv(r'C:\Users\Priyadharshini\Desktop\Final project data.csv')
```

```python
df = df.drop(columns=['patient_id'])
```

```python
df.fillna(df.median(numeric_only=True), inplace=True)
```

```python
df = pd.get_dummies(df, columns=['gender'], drop_first=True)
```

```python
# Separate features & targets

X = df.drop(columns=['outcome', 'length_of_stay_days'])
y_class = df['outcome']
y_reg = df['length_of_stay_days']
```

```python
# Feature Scaling

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

clean_df = pd.DataFrame(X_scaled, columns=X.columns)
clean_df['outcome'] = y_class.values
clean_df['length_of_stay_days'] = y_reg.values

```

```python
clean_df.to_csv("cleaned_data.csv", index=False)
```
